## 9.5 Yield curve*

Plot the zero-coupon yield curve, that is, the yields on zero-coupon bonds as a function of their time to maturity. Include the overnight interest rate of 3.7\%.

```python
import matplotlib.pyplot as plt
# to format the y-axis in percent
import matplotlib.ticker as mtick
```

```python
# time to maturity in years
maturities = [0, 1, 2, 3]
```

### Initial yield curve

```python
# yields in percent
yields = [3.7, 3.82, 4.60, 3.92]
```

```python
# include a dotted line
x_dotted = (maturities[1], maturities[3])
y_dotted = (yields[1], yields[3])
```

```python
fig, ax = plt.subplots()
ax.plot(maturities, yields, marker="o", label="Initial")
ax.plot(x_dotted, y_dotted, linestyle="dotted", label="Interpolated")
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.ylim(0, 5)
plt.grid()
plt.legend(loc="lower left")
plt.title("Zero-coupon yield curve")
plt.ylabel("Yield")
plt.xlabel("Time to maturity in years")
plt.savefig("yield_curve.pdf")
plt.show()
```

## 9.7 Yield curve trading*

### Level change up

```python
# time to maturity in years
maturities_new = [1, 2, 3]
# yields in percent
yields_new = [4.82, 4.87, 4.92]
```

```python
fig, ax = plt.subplots()
ax.plot(maturities, yields, marker="o", label="Initial")
ax.plot(x_dotted, y_dotted, linestyle="dotted", label="Interpolated")
ax.plot(maturities_new, yields_new, marker="o", label="Level change up")
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.ylim(0, 6)
plt.grid()
plt.legend(loc="lower left")
plt.title("Zero-coupon yield curve")
plt.ylabel("Yield")
plt.xlabel("Time to maturity in years")
plt.savefig("yield_curve_level_up.pdf")
plt.show()
```
